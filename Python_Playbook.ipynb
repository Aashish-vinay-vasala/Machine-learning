```python
name = input("Enter file: : ")
handle = open(name)
text = handle.read()
words = text.split()

counts = dict()
for word in words:
  counts[word] = counts.get(word,0) + 1

bigcount = None
bigword = None
for word,count in counts.items():
  if bigcount is None or count > bigcount:
    bigword = word
    bigcount = count

print(bigword,bigcount)
```

```python
fhand = open('essay.txt')
counts = dict()
for line in fhand:
  words = line.split()
  for word in words:
    counts[word] = counts.get(word,0) + 1
print(counts)
```

```python
lst = list()
for key,val in counts.items():
  lst.append((val,key))
```

```python
lst.sort(reverse=True)
```

```python
for val,key in lst[:10]:
  print(key,val)
```

```python
fhand = open('mbox-short.txt')
counts = dict()
for line in fhand:
  words = line.split()
  for word in words:
    counts[word] = counts.get(word,0) + 1
lst = list()
for key,val in counts.items():
  lst.append((val,key))
print(lst)
```

```python
lst.sort(reverse=True)
for val,key in lst[:10]:
  print(key,val)
```

```python
##Using re.search() like StartsWith()

import re

hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('From:',line):
      print(line)
```
