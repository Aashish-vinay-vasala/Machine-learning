```python
url1 = 'https://raw.githubusercontent.com/google/dspl/master/samples/google/canonical/countries.csv'
req1 = requests.get(url1)
df1 = StringIO(req1.text)

l = []
for x in range(10):
  l.append(x)

print(l)

df2 = pd.read_csv(df1,skiprows=l)
df2
```

#11.Converters

```python
def rename(name):
  if name == 'Dark Green':
    return 'DG'
  else:
    return name
```

```python
rename('Dark Green')
```

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/prm2711/python-recommender/master/zomato.csv"

req2 = requests.get(url)
data2 = StringIO(req2.text)

pd.read_csv(data2,converters={'Rating color':rename}).head()
```

#13.Loading huge dataset in chunks

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/prm2711/python-recommender/master/zomato.csv"

req2 = requests.get(url)
data2 = StringIO(req2.text)

data2 = pd.read_csv(data2,chunksize=100)
```

```python
for chunk in data2:
  print(chunk.shape)
```
